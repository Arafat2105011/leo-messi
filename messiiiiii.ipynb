# Top 20 Favourite Opponents graph


```python


top_opponents = data['Opponent'].value_counts().head(20)


plt.figure(figsize=(12, 6))
sns.barplot(x=top_opponents.values, y=top_opponents.index, orient='h')
plt.title('Top 20 Oponents')
plt.xlabel('Number of goals')
plt.ylabel('Opponents')
plt.show()
```

# Top assist providers graph

```python

top_assisters = data['Goal_assist'].value_counts().head(15)


plt.figure(figsize=(15, 8))
sns.barplot(x=top_assisters.values, y=top_assisters.index, orient='h')
plt.title('Top 15 assister')
plt.xlabel('Assisters')
plt.ylabel('number of assist')
plt.show()
#0=[no assists]
```

# Top Favourite 15 minutes barplotting

```python
top_15_favourite_minutes = data['Minute'].value_counts().head(15)

plt.figure(figsize=(19,8))
sns.barplot(x=top_15_favourite_minutes.values, y=top_15_favourite_minutes.index, orient='h')
plt.title('"GOATS favourite minutes"')
plt.xlabel('Value of Minutes')
plt.ylabel('Minutes')
plt.show()
```

```python
plt.figure(figsize=(14,8))
sns.countplot(x='Competition',data=data , order=data['Competition'].value_counts().index)
plt.title('Favourite Competition')
plt.xticks(rotation=70)
plt.show()
```
